### Question 2

##### Plot the distribution of failed orders by hours. Is there a trend that certain hours have an abnormally high proportion of one category or another? What hours are the biggest fails? How can this be explained?

```python
df["order_hour"] = df["order_time"].str.split(":").apply(lambda split:split[0])
```

```python
df1 =df.groupby("order_hour")["order_gk"].count()

df1.plot(figsize=(7,7), legend = True, xticks = range(0,24), title = "Count of Failed orders by Hour of Day")
plt.show()
```

```python
df2 =df.groupby(["order_hour","is_driver_assigned","order_status"])["order_gk"].count()
df2
```

```python
df2.reset_index().pivot(index= "order_hour", columns = ["is_driver_assigned","order_status"],values = "order_gk").plot(xticks=range(0, 24),
                                                           figsize=(13, 7),
                                                           title="Count of Failed Orders Per Hour and Category")
```
